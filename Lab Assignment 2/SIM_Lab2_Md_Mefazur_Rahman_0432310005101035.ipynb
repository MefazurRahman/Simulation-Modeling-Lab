# Task-2: Analyze Matrix A
Using NumPy, compute:


*   Shape of A
*   Determinant

*   Rank
*   Eigenvalues

*   Inverse (only if determinant ≠ 0)












```python
A = np.array([[5+2,3+1],[2*5,3+2]])

# Shape of matrix A
shape = A.shape
print("Shape of matrix A is: ",shape)

# Determinent of A
det_A = np.linalg.det(A)
print("Determinent of matrix A is: ",det_A)

# Rank of matrix A
rank_A = np.linalg.matrix_rank(A)
print("Rank of matrix A is : ",rank_A)

# Eigen values
eigen_values,eigen_vectors = np.linalg.eig(A)
print("Eigenvalues : ",eigen_values)

# Inverse of  matrix A
if np.linalg.det(A)!=0:
  inv_A = np.linalg.inv(A)
  print("Inverse of matrix A is :\n",inv_A)
else: print("The determinent of the matrix is 0")
```
