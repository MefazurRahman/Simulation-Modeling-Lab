```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

# Assignment 1: Basic Matrix Operations

```python
### 1.Create two 3Ã—3 matrices with random integers in Python.
matA = np.random.randint(0, 10, size=(3, 3))
matB = np.random.randint(0, 20, size=(3, 3))

print("Matrix A:\n", matA)
print("\nMatrix B:\n", matB)

### 2.Perform addition, subtraction, and multiplication operations on these matrices.
matrix_sum = matA + matB
matrix_diff = matA - matB
matrix_prod = np.dot(matA, matB)
print("\nAddition: \n", matrix_sum)
print("\nSubtraction: \n", matrix_diff)
print("\nMultiplication: \n", matrix_prod)

### 3.Compute the determinant, inverse, and rank of one of the matrices (Matrix-A).
det_A = np.linalg.det(matA)
if np.linalg.det(matA) != 0:
    inv_A = np.linalg.inv(matA)
else:
    inv_A = "Cannot compute inverse matrix, because the determinant is 0"

rank_A = np.linalg.matrix_rank(matA)
print("\nDeterminant of A:", det_A)
print("\nInverse of A:\n", inv_A)
print("\nRank of A:", rank_A)
```

# Assignment 2: Matrix Plotting and Visualization

1. Generate and Plot Two Random Vectors:

```python
### a.Create two vectors, each containing 15 random floats of your choice.
vec1 = np.random.rand(15)
vec2 = np.random.rand(15)

print("x vector: \n",vec1)
print("\n y vector: \n",vec2)

### b.Plot these vectors on the same graph using matplotlib to compare how they look.
plt.scatter(vec1, vec2, color='b', label='Random Points')

### c.Label the axes appropriately and give the plot a title.
plt.title('Scatter Plot of Two Random Vectors (vec1&vec2)')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend()
plt.grid(True)
plt.show()
```

2. Generate and Visualize a Random Matrix:

```python
### a.Create a 4x4 matrix with random values of your choice.
matrix = np.random.randint(0, 10, size=(4, 4))
print(matrix, "\n")

### b.Display the matrix using a heatmap with appropriate color mapping (e.g. seaborn or matplotlib).
sns.heatmap(matrix, annot=True, cmap='coolwarm', linewidths=0.5, linecolor ='black')
plt.title('Heatmap of 4x4 Matrix')
plt.show()

print("\n")

### c.Add labels to the heatmap to indicate row and column numbers.
sns.heatmap(matrix, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black',
            xticklabels=[1, 2, 3, 4],
            yticklabels=[1, 2, 3, 4])
plt.title('Heatmap of 4x4 Random Matrix with Labels')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```

3. Matrix Operations and Visualization:

```python
### a.Create two 4x4 matrices with random values of your choice.
matA = np.random.randint(0, 10, size=(4, 4))
matB = np.random.randint(0, 20, size=(4, 4))

print("Matrix A:\n", matA)
print("\nMatrix B:\n", matB)

### b.Perform the following operations: Addition, Subtraction, Multiplication
matrix_sum = matA + matB
matrix_diff = matA - matB
matrix_prod = np.dot(matA, matB)
print("\nAddition: \n", matrix_sum)
print("\nSubtraction: \n", matrix_diff)
print("\nMultiplication: \n", matrix_prod)

### c.For each of these operations, visualize the result using bar plots. Each bar should represent one element of the matrix.
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].bar(range(1, 17), matrix_sum.flatten(), color='r')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Element Index')
axes[0].set_ylabel('Value')
axes[1].bar(range(1, 17), matrix_diff.flatten(), color='g')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Element Index')
axes[1].set_ylabel('Value')
axes[2].bar(range(1, 17), matrix_prod.flatten(), color='b')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Element Index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
